```python
## here we write a function that parse the json data into the data frame
def parseJSONdataToDataframe(json_data):
    j = json.loads(json_data)
    
    try:
        updateTimeDtaa = j['updateTime']
        waitTimeData = j['waitTime']
        
        df_data = {
            "hospName":[],
            "topWait":[],
            "updateTime":[],
        }

        updated_df = pd.DataFrame(df_data, columns=['hospName', 'topWait', 'updateTime'])

        hospName_list = []
        topWait_list = []
        
        for i in range(len(waitTimeData)):
            hospName_list.append(waitTimeData[i]['hospName'])
            topWait_list.append(waitTimeData[i]['topWait'])

        updated_df['hospName'] = hospName_list
        updated_df['topWait'] = topWait_list
        updated_df['updateTime'] = j['updateTime']

        print('Showing the first 5 rows\n')
        print(updated_df.head())
        return updated_df

    except Exception:
        pass

```

```python
# This function is to get ANE data of certain range of datetime from the api of data.gov.hk

def ANE_data_to_csv(range_datetime):
    print('Getting data from data.gov.hk...\n\n')

    api_json_data = getJSONfromAPI(range_datetime)
    updated_df = parseJSONdataToDataframe(api_json_data)
    
    # check if the file exist
    if os.path.isfile('waitTimeData.csv'):
        
        exist_df = pd.read_csv("waitTimeData.csv")

        exist_df = exist_df.append(updated_df)

        exist_df.to_csv(r'waitTimeData.csv', index = None, header=True)

    else:
        updated_df.to_csv(r'waitTimeData.csv', index=None, header=True)


```

```python
datetime_period = full_range[0:10]

for dt in datetime_period:
    ANE_data_to_csv(dt)

```

## Gradient Boosting Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.7)

print(len(X_train), len(y_train), len(X_test), len(y_test))
```

```python
# importing libraries for making pipelines
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

```

```python
pipelines ={
    'rf': make_pipeline(StandardScaler(), RandomForestClassifier(random_state=123)),
    'gb': make_pipeline(StandardScaler(), GradientBoostingClassifier(random_state=123))
}
```

```python
rf_hyperparameters = {
    'randomforestclassifier__n_estimators': [100,200],
    'randomforestclassifier__max_features': ['auto', 'sqrt', 0.33]
}
```

```python
gb_hyperparameters = {
    'gradientboostingclassifier__n_estimators': [100, 200],
    'gradientboostingclassifier__learning_rate':[0.05, 0.1, 0.2],
    'gradientboostingclassifier__max_depth': [1,3,5]
}
```

```python
hyperparameters = {
    'rf': rf_hyperparameters,
    'gb': gb_hyperparameters
}
```

```python
for name, pipeline in pipelines.items():
    print(name)
    print(pipeline)
```

```python
fitted_model = {}
for name, pipeline in pipelines.items():
    model = GridSearchCV(pipeline, hyperparameters[name], n_jobs=-1)
    
    model.fit(X_train, y_train)

    fitted_model[name] = model

    print(name, 'has been fitted')
```

```python
pred_rf = fitted_model['rf'].predict(X_test)
```

```python
pred_gb = fitted_model['gb'].predict(X_test)
```
